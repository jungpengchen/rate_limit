```python

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You're a good stroy teller, and you should tell me a long story, should over 200 words."),
        MessagesPlaceholder(variable_name="history"),
        ("human", "{question}"),
    ]
)

chain = prompt | AzureChatOpenAI(
    azure_endpoint=os.environ["azure_endpoint"],
    deployment_name=os.environ["deployment_name"],
    openai_api_version=os.environ["openai_api_version"],
    openai_api_key=os.environ["openai_api_key"],
    
)

chain_with_history = RunnableWithMessageHistory(
    chain,
    lambda session_id: RedisChatMessageHistory(
        session_id, url="redis://localhost:6379"
    ),
    input_messages_key="question",
    history_messages_key="history",
)

config = {"configurable": {"session_id": "orin_history3"}}

def ask_math_question(question: str):
    print(chain_with_history.invoke({"question": question}, config=config).content)
```

# Chatbot with RAG and store memory in Redis

```python
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader("https://en.wikipedia.org/wiki/Reformation")
data = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
all_splits = text_splitter.split_documents(data)
```

```python
from langchain_community.vectorstores import Chroma
from langchain_openai import OpenAIEmbeddings

vectorstore = Chroma.from_documents(documents=all_splits, embedding=OpenAIEmbeddings(api_key=API_KEY))
retriever = vectorstore.as_retriever(k=4)
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain.chains.history_aware_retriever import create_history_aware_retriever
from langchain.chains import create_retrieval_chain


chat = AzureChatOpenAI(
    azure_endpoint=os.environ["azure_endpoint"],
    deployment_name=os.environ["deployment_name"],
    openai_api_version=os.environ["openai_api_version"],
    openai_api_key=os.environ["openai_api_key"],
    
)



SYSTEM_TEMPLATE = """
You are a history teacher, and you should base on context to answer the question to me,

At the beginning, you should ask my name, and follow conversation, you should call my name first, and then answer the question.

You can chat to me about history, and I will ask you questions about history.

If the context doesn't contain any relevant information to the question, don't make something up and just tell me you don't know.:

<context>
{context}
</context>
"""

question_answering_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            SYSTEM_TEMPLATE,
        ),
        MessagesPlaceholder(variable_name="chat_history"),
        ("human", "{input}"),
    ]
)
document_chain = create_stuff_documents_chain(chat, question_answering_prompt)


prompt = ChatPromptTemplate.from_messages(
    [
        MessagesPlaceholder(variable_name="chat_history"), 
     ("human", "{input}"),
     ("human","Given the above conversation, generate a search query to look up in order to get information relevant to the conversation. Only respond with the query, nothing else.")
     ])
        

history_aware_retriever = create_history_aware_retriever(llm=chat, retriever=retriever,prompt=prompt)


test_chain = create_retrieval_chain(retriever=history_aware_retriever,combine_docs_chain=document_chain)

test_chain.invoke({"input": "What is reformation?","chat_history":[]})


history_chain_with_memory = RunnableWithMessageHistory(
    test_chain,
    lambda session_id: RedisChatMessageHistory(
        session_id, url="redis://localhost:6379"
    ),
    input_messages_key="input",
    output_messages_key="answer",
    history_messages_key="chat_history",
)

config = {"configurable": {"session_id": "orin_history_teaching_8"}}

def ask_question(question: str):
    print(history_chain_with_memory.invoke( {
        # "context": docs,
        "input": question,
        
    }, config=config)["answer"])
```
